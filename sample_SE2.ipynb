```python
# This notebook is concerned with working with SE(2) transforms.
# The code defines two functions:
# – The function `sample_se2` generates a random 3×3 homogeneous
#   transformation matrix whose rotation angle θ is drawn from
#   [0, θ0] and whose x,y translations are drawn from [a, b].
#   It returns the sampled θ, x, y and the corresponding matrix G.
# – The function `is_se2` takes a 3×3 array and verifies that it
#   represents a member of SE(2): the last row must be [0,0,1],
#   the top‑left 2×2 block must be orthogonal with determinant 1.
```

```python
import numpy as np
```

```python
def sample_se2(theta0: float, a: float, b: float):
    """
    Sample an SE(2) matrix with:
      theta in [0, theta0]
      x, y in [a, b]
    """
    if theta0 < 0:
        raise ValueError("theta0 must be >= 0.")
    if a > b:
        raise ValueError("Need a <= b.")

    theta = np.random.uniform(0.0, theta0)
    x = np.random.uniform(a, b)
    y = np.random.uniform(a, b)

    c = np.cos(theta)
    s = np.sin(theta)

    G = np.array([
        [c, -s, x],
        [s,  c, y],
        [0.0, 0.0, 1.0]
    ])

    return theta, x, y, G
```

```python
def is_se2(G: np.ndarray, tol: float = 1e-10) -> bool:
    """
    Check whether a 3x3 matrix is in SE(2).
    """
    if G.shape != (3, 3):
        return False

    # Last row must be [0,0,1]
    if not np.allclose(G[2], np.array([0.0, 0.0, 1.0]), atol=tol):
        return False

    R = G[:2, :2]

    # Check orthogonality
    if not np.allclose(R.T @ R, np.eye(2), atol=tol):
        return False

    # Check determinant
    if not np.isclose(np.linalg.det(R), 1.0, atol=tol):
        return False

    return True
```

```python
theta, x, y, G = sample_se2(theta0=np.pi/3, a=-2.0, b=2.0)

print("theta =", theta)
print("x =", x)
print("y =", y)
print("G =\n", G)
print("Is SE(2)?", is_se2(G))
```
